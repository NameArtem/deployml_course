## PDB: Intro to `pdb`

PDB нам поможет.

Этим инструментом очень легко пользоваться


```python
import pdb; pdb.set_trace()
```

Метод [`set_trace()`](https://docs.python.org/3/library/pdb.html#pdb.set_trace) специальная точка, с которой начинается тестирование, её надо поместить в код (`main.py`):

`file: main.py` 
```python
from dicegame.runner import GameRunner


def main():
    print("Определите значение в игровых кубиках")
    print("Начем")
    import pdb; pdb.set_trace() # add pdb here
    GameRunner.run()


if __name__ == "__main__":
    main()
```

Запустим заново `main.py`

```shell
python main.py
```
```
Определите значение в игровых кубиках
Начем
> .../main.py(9)main()
-> GameRunner.run()
(Pdb) 
```

Начен тестировать


## 5 `pdb` команд

Библиотека имеет свои управляющие команды

1. `l(ist)` - показывает следующую строку, которая выполняется и 5 строк до и после неё (11 строчек всегр)
2. `s(tep)` - сделает 1 шаг выполнения, до следуюшей возможной остановки
3. `n(ext)` - выполнение кода до следующей строчки
4. `b(reak)` - остановка
5. `r(eturn)` - выполнение кода до следующего return условия
6. `c(ontinue)` -  выполнение до ошибки


**NB**: не забываем про помоника:
`h(elp) - выводит список аргументов `

### 4. `b(reak)` 

```
b(reak) [ ([filename:]lineno | function) [, condition] ]
    Without argument, list all breaks.

    With a line number argument, set a break at this line in the
    current file.  With a function name, set a break at the first
    executable line of that function.  If a second argument is
    present, it is a string specifying an expression which must
    evaluate to true before the breakpoint is honored.

    The line number may be prefixed with a filename and a colon,
    to specify a breakpoint in another file (probably one that
    hasn't been loaded yet).  The file is searched for on
    sys.path; the .py suffix may be omitted.
```

### `pdb` Post Mortem

```

    Enter post-mortem debugging of the given traceback object. If no traceback is given, it uses the one of the exception that is currently being handled

pdb.pm()
    Enter post-mortem debugging of the traceback found in sys.last_traceback.
```
